在 Python 中，变量就像一个贴着标签的容器，你可以用它来存储数据。你不需要提前声明变量的类型，只需要给它赋一个值，Python 会自动判断它的类型。

### 1\. 变量的创建和赋值

创建变量非常简单，你只需要想一个变量名，然后使用等号 `=` 将一个值赋给它。

`变量名 = 值`

```python
# 创建一个名为 message 的变量，并存储文本（字符串）
message = "你好, 世界！"
print(message)

# 创建一个名为 number 的变量，并存储一个整数
number = 100
print(number)

# 创建一个名为 pi 的变量，并存储一个小数（浮点数）
pi = 3.14159
print(pi)

# 创建一个名为 is_learning 的变量，并存储一个布尔值 (True 或 False)
is_learning = True
print(is_learning)
```

**输出:**

```
你好, 世界！
100
3.14159
True
```

### 2\. 变量的使用

你可以随时使用变量名来获取它存储的值，并用在计算、打印等操作中。

```python
# 使用变量进行数学运算
x = 10
y = 5
sum_result = x + y

print("x 的值是:", x)
print("y 的值是:", y)
print("x 和 y 的和是:", sum_result)

# 使用变量连接字符串
first_name = "张"
last_name = "三"
full_name = first_name + last_name
print("完整的姓名是:", full_name)
```

**输出:**

```
x 的值是: 10
y 的值是: 5
x 和 y 的和是: 15
完整的姓名是: 张三
```

### 3\. 更新变量的值

你可以随时给一个已经存在的变量赋一个新的值，这会覆盖掉它之前存储的值。

```python
# 创建变量
price = 50
print("这件T恤的初始价格是:", price)

# 更新价格（打折）
price = 40
print("打折后，现在的价格是:", price)
```

**输出:**

```
这件T恤的初始价格是: 50
打折后，现在的价格是: 40
```

### 4\. 查看变量的类型

如果你不确定一个变量存储的是哪种类型的数据，可以使用 `type()` 函数来查看。

```python
greeting = "Hello"
count = 5
pi_value = 3.14
is_active = False

print(greeting, "的类型是:", type(greeting)) # str 代表字符串
print(count, "的类型是:", type(count))       # int 代表整数
print(pi_value, "的类型是:", type(pi_value)) # float 代表浮点数
print(is_active, "的类型是:", type(is_active)) # bool 代表布尔值
```

**输出:**

```
Hello 的类型是: <class 'str'>
5 的类型是: <class 'int'>
3.14 的类型是: <class 'float'>
False 的类型是: <class 'bool'>
```

### 5\. 变量的命名规则和最佳实践

给变量起一个好名字非常重要，这能让你的代码更容易阅读和理解。

  * **规则（必须遵守）**:

      * 变量名只能包含字母、数字和下划线 (`_`)。
      * 变量名不能以数字开头。
      * 变量名不能是 Python 的关键字（比如 `if`, `for`, `class` 等）。
      * 变量名是区分大小写的（`name` 和 `Name` 是两个不同的变量）。

  * **最佳实践 (建议遵守)**:

      * 使用小写字母和下划来分隔单词，这种风格被称为 **snake\_case**。例如：`my_first_name`。
      * 变量名应该具有描述性，让人一眼就能看出它存储的是什么。例如，用 `user_name` 而不是 `un`。

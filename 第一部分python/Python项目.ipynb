

## 🛒 **校园小卖部管理系统**

### 项目描述：
创建一个模拟校园小卖部的购物系统，包含商品管理、购物车、结账、会员积分等功能。

### 核心功能要求：

1. **商品管理**
   - 商品分类（零食、饮料、文具、日用品）
   - 添加/删除/修改商品信息
   - 库存管理和提醒
   - 商品搜索功能

2. **购物功能**
   - 浏览商品列表
   - 加入购物车
   - 修改购物车（增减数量、删除商品）
   - 实时计算总价

3. **会员系统**
   - 用户注册/登录
   - 会员等级（普通、银卡、金卡）
   - 积分累计和兑换
   - 购物历史记录

4. **优惠活动**
   - 满减优惠（满100减10）
   - 会员折扣（银卡9折、金卡8折）
   - 限时特价商品
   - 买二送一活动

5. **结账系统**
   - 多种支付方式（现金、积分、混合支付）
   - 打印小票
   - 找零计算
   - 订单编号生成

6. **数据统计**
   - 热销商品排行
   - 日/周/月销售额统计
   - 库存预警列表
   - 会员消费排行

### 技术要求覆盖：
- **面向对象**：Product（商品）、User（用户）、Cart（购物车）、Order（订单）类
- **文件操作**：保存商品信息、用户数据、订单记录（JSON/CSV格式）
- **数据结构**：字典（商品库）、列表（购物车）、集合（商品分类）
- **异常处理**：库存不足、支付错误、输入验证
- **函数编程**：计算折扣、生成报表、数据筛选
- **时间处理**：订单时间、特价时段、统计周期
